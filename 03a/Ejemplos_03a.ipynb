```python
import cv2 as cv

# Cargar la imagen
fichero = 'images/Einstein.png' #Imagen en blanco y negro
#fichero = 'images/eii.png'     #Imagen en color
imagen = cv.imread(fichero)

alto, ancho, canales = imagen.shape

# Nombre del archivo de salida
nombre_video = 'salida.mp4'
#Tasa de fotogramas. 
#Si quieres que el vídeo sea más rápido, aumenta el valor de fps. 
#Si quieres que sea más lento, disminuye el valor de fps.
fps = 30.0 

# Definir el codec y crear el objeto VideoWriter
fourcc = cv.VideoWriter_fourcc(*'mp4v')  # Codec para .mp4
video = cv.VideoWriter(nombre_video, fourcc, fps, (ancho, alto))

# Añadir cada imagen al vídeo
frame = imagen
for i in range(0,alto):
    for j in range(0,ancho):
       frame[i,j] = 0 
       video.write(frame)

# Liberar el objeto VideoWriter
video.release()

print(f"Vídeo guardado como {nombre_video}")
```

```python
# Cálculo de histogramas
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

image = cv.imread('images/dog1.png')
gray = cv.cvtColor(image, cv.COLOR_BGR2GRAY)

cv.imshow('Color', image)
cv.imshow('Gris', gray)

# histograma de niveles de grises
gray_hist = cv.calcHist([gray], [0] , None,[256], [0,256])
plt.figure()
plt.title('Histograma en escala de grises')
plt.xlabel('Bins')
plt.ylabel('# de pixeles')
plt.plot(gray_hist)
plt.savefig("gray_histogram.svg")
plt.show()

# histograma de color
colors = ('b','g','r')
for i, col in enumerate(colors):
    hist = cv.calcHist([image], [i], None, [256], [0,256])

    #imágenes: lista de imágenes en forma de matrices numpy. Todas las imágenes deben tener el mismo tipo de datos y el mismo tamaño.
    #canales: lista de los canales utilizados para calcular los histogramas.
    #máscara: máscara opcional (matriz de 8 bits) del mismo tamaño que la imagen de entrada.
    #histSize: tamaños de histogramas en cada dimensión
    #rangos: Matriz de matrices de dimensiones de los límites de los contenedores del histograma en cada dimensión
    #hist: histograma de salida
    #acumular: bandera de acumulación, permite calcular un solo histograma a partir de varios conjuntos de matrices.

    plt.plot(hist, color=col)
    plt.xlim([0,256])
plt.title('Histograma en color')
plt.xlabel('Bins')
plt.ylabel('# de pixeles')
plt.savefig("color_histogram.svg")
plt.show()

cv.waitKey(0)
cv.destroyAllWindows()
```

```python
# Ecualización de histogramas
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

image = cv.imread('images/Ecualizacion3.png') #'images/Ecualizacion1.png'
gray = cv.cvtColor(image, cv.COLOR_BGR2GRAY)

cv.imshow('Original', image)
cv.imshow('Original en gris', gray)

# histograma de niveles de grises original
gray_hist = cv.calcHist([gray], [0] , None,[256], [0,256])
gray_cum = np.cumsum(gray_hist)
plt.figure()
plt.title('Histograma de grises')
plt.xlabel('Bins')
plt.ylabel('# de pixeles')
plt.plot(gray_hist/np.max(gray_hist))
plt.plot(gray_cum/np.max(gray_cum))
plt.legend(['Histograma', 'Histograma Acumulado'],loc='right')
plt.savefig("gray_histogram.svg")
plt.show()

# ecualización de niveles de grises
gray_equalized = cv.equalizeHist(gray)
cv.imshow('Ecualizado de grises', gray_equalized)

# histograma de niveles de grises ecualizado
gray_hist = cv.calcHist([gray_equalized], [0] , None,[256], [0,256])
gray_cum = np.cumsum(gray_hist)
plt.figure()
plt.title('Histograma ecualizado de grises')
plt.xlabel('Bins')
plt.ylabel('# de pixeles')
plt.plot(gray_hist/np.max(gray_hist))
plt.plot(gray_cum/np.max(gray_cum))
plt.legend(['Histograma', 'Histograma Acumulado'],loc='right')
plt.savefig("gray_histogram_equalized.svg")
plt.show()

# histograma de imagen a color
colors = ('b','g','r')
for i, col in enumerate(colors):
    hist = cv.calcHist([image], [i], None, [256], [0,256])
    hist_cum = np.cumsum(hist)
    plt.plot(hist/np.max(hist), color=col)
    plt.plot(hist_cum/np.max(hist_cum), color=col)
    plt.xlim([0,256])
plt.title('Histograma de colores')
plt.xlabel('Bins')
plt.ylabel('# de pixeles')
plt.savefig("color_histogram.svg")
plt.show()

# ecualización e histograma de imagen a color
image_equalized = image
for i, col in enumerate(colors):
    image_equalized[:,:,i] = cv.equalizeHist(image[:,:,i])
    
    hist = cv.calcHist([image_equalized], [i], None, [256], [0,256])
    hist_cum = np.cumsum(hist)
    plt.plot(hist/np.max(hist), color=col)
    plt.plot(hist_cum/np.max(hist_cum), color=col)
    plt.xlim([0,256])
plt.title('Histograma ecualizado de colores')
plt.xlabel('Bins')
plt.ylabel('# of pixels')
plt.savefig("color_histogram_equalized.svg")
plt.show()

cv.imshow('Ecualizado de color', image_equalized)

cv.waitKey(0)
cv.destroyAllWindows()
```

```python
# Función para añadir ruido Gaussiano a una imagen o video

import cv2 as cv
import numpy as np

def add_gaussian_noise(image, mean=0.0, sigma=30):
    noisy = image + np.random.normal(mean,sigma,image.shape)
    noisy[noisy>255]=255
    noisy[noisy<0]=0
    return noisy.astype(np.uint8)

#image = cv.imread('images/manzana.png')
video = cv.VideoCapture('videos/times-square.mp4')
#video = cv.VideoCapture(0)

state = True
while state:
    state, image = video.read()
    noisy_image = add_gaussian_noise(image)
    cv.imshow('Video ruidoso', noisy_image)
    if cv.waitKey(20)==27:
        state = False
    
video.release()
#cv.waitKey(0)
cv.destroyAllWindows()
```
